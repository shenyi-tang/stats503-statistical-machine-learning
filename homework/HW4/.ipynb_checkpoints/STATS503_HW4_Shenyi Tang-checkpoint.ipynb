#### Quesiton 1

**ISLP chapter 4, conceptual exercise #2**

**It was stated in the text that classifying an observation to the class for which (4.17) is largest is equivalent to classifying an observation to the class for which (4.18) is largest. Prove that this is the case. In other words, under the assumption that the observations in the $k$th class are drawn from a $N(u_k,\sigma^2)$ distribution, the Bayes classifier assigns an observation to the class for which the discriminant function is maximized.**

That is to say， maximizing
$$
\begin{equation}
p_k(x)=\cfrac{\pi_k\frac{1}{\sqrt{2\pi}\sigma}\text{exp}(-\frac{1}{2\sigma^2}(x-\mu_k)^2)}{\sum_{l=1}^{K}{\pi_l\frac{1}{\sqrt{2\pi}\sigma}\text{exp}(-\frac{1}{2\sigma^2}(x-\mu_l)^2)}}
\end{equation}
$$
is equivalent to maximizing
$$
\begin{equation}
\delta_k(x)=x\cdot\cfrac{\mu_k}{\sigma^2}-\cfrac{\mu_k^2}{2\sigma^2}+\text{log}(\pi_k)
\end{equation}
$$
For each $p_k(x)$, there are only differences in numerator, that is to say, we have to find the largest numerator $\pi_k\frac{1}{\sqrt{2\pi}\sigma}\text{exp}(-\frac{1}{2\sigma^2}(x-\mu_k)^2)$ among all the $k$. 

To simplify the process, we can take $\text{log}$ to the numerator of $p_k(x)$, we get 
$$-\cfrac{x^2}{2\sigma^2}+x\cdot\cfrac{\mu_k}{\sigma^2}-\cfrac{\mu_k^2}{2\sigma^2}+\text{log}(\cfrac{1}{\sqrt{2\pi}\sigma})+\text{log}{\pi_k}$$.

As we only care about $k$-related items, we finally get $x\cdot\cfrac{\mu_k}{\sigma^2}-\cfrac{\mu_k^2}{2\sigma^2}+\text{log}(\pi_k)$, which is $\delta_k$

#### Quesiton 2

**ISLP chapter 4, conceptual exercise #3**

**This problem relates to the QDA model, in which the observations within each class are drawn from a normal distribution with a class-specific mean vector and a class specific covariance matrix. We consider the simple case where $p = 1$; i.e. there is only one feature. Suppose that we have $K$ classes, and that if an observation belongs to the $k$th class then X comes from a one-dimensional normal distribution, $X\sim N(µ_k , \sigma^2_k)$. Recall that the density function for the one-dimensional normal distribution is given in (4.16). Prove that in this case, the Bayes classifier is not linear. Argue that it is in fact quadratic.**

For the cases that variances are different in different classes, we have
$$
\begin{equation}
p_k(x)=\cfrac{\pi_k\frac{1}{\sqrt{2\pi}\sigma_k}\text{exp}(-\frac{1}{2\sigma^2_k}(x-\mu_k)^2)}{\sum_{l=1}^{K}{\pi_l\frac{1}{\sqrt{2\pi}\sigma_l}\text{exp}(-\frac{1}{2\sigma_l^2}(x-\mu_l)^2)}}
\end{equation}
$$
Similarly, the denominators are equal for each $p_k(x)$ and we need to find the largest numerator.

Taking $\text{log}$ to the numerator of $p_k(x)$ and cancel the items that are not related with $k$, we can have 
$$
-\cfrac{x^2}{2\sigma^2_k}+x\cdot\cfrac{\mu_k}{\sigma^2_k}-\cfrac{\mu_k^2}{2\sigma^2_k}+\text{log}(\cfrac{1}{\sqrt{2\pi}\sigma_k})+\text{log}{\pi_k}
$$
which is quadratic

**(b) Fit the LDA model using a training data period from 2001 to 2004, with `Direction` as the response and `Lag1` and `Lag2` as predictors. Use the model to estimate the Bayes optimal predictions on the held out test data (that is, the data from 2005). Compute the confusion matrix and the overall fraction of correct predictions for the test data.**

```python

train = smarket[smarket["Year"] != 2005][["Direction", "Lag1", "Lag2"]]
y_train = train["Direction"]
x_train = train[["Lag1", "Lag2"]]

test = smarket[smarket["Year"] == 2005][["Direction", "Lag1", "Lag2"]]
y_test = test["Direction"]
x_test = test[["Lag1", "Lag2"]]

# mu
mu_train = train.groupby("Direction")[["Lag1", "Lag2"]].mean()

# covariance matrix
sigma = np.cov(x_train.T)

# priors
pi = dict(y_train.value_counts() / len(y_train))

# LDA function
def lda_score(x, mu, sigma, pi):
    sigma_inv = np.linalg.inv(sigma)
    scores = {}

    for direction in ['Up', 'Down']:
        mu_k = mu.loc[direction].values
        prior = np.log(pi[direction])

        score = (x @ sigma_inv @ mu_k - 0.5 * mu_k.T @ sigma_inv @ mu_k + prior)
        scores[direction] = score

    return max(scores, key=scores.get)


y_pred = x_test.apply(lambda row: lda_score(row, mu_train, sigma, pi), axis=1)

# confussion matrix and 
conf_matrix = confusion_matrix(y_test, y_pred)
accuracy = accuracy_score(y_test, y_pred)

print("\nConfusion Matrix:")
print(pd.DataFrame(conf_matrix, 
                   index=['Actual Down', 'Actual Up'], 
                   columns=['Predicted Down', 'Predicted Up']))
print(f"\nAccuracy: {accuracy:.4f}")
```

**(c) Repeat (b) using QDA.**

```python
train = smarket[smarket["Year"] != 2005][["Direction", "Lag1", "Lag2"]]
y_train = train["Direction"]
x_train = train[["Lag1", "Lag2"]]

test = smarket[smarket["Year"] == 2005][["Direction", "Lag1", "Lag2"]]
y_test = test["Direction"]
x_test = test[["Lag1", "Lag2"]]

# mu
mu_train = train.groupby("Direction")[["Lag1", "Lag2"]].mean()

# priors
pi = dict(y_train.value_counts() / len(y_train))

# new covariance matrix
sigma_qda = {label: np.cov(x_train[y_train == label].T) for label in ['Up', 'Down']}

# qda function
def qda_score(x, mu, sigma, pi):
    scores = {}

    for direction in ['Up', 'Down']:
        mu_k = mu.loc[direction].values
        sigma_k = sigma[direction]
        sigma_k_inv = np.linalg.inv(sigma_k)
        det_sigma_k = np.linalg.det(sigma_k)

        # 计算 QDA 评分
        term1 = -0.5 * np.log(det_sigma_k)
        term2 = -0.5 * (x - mu_k).T @ sigma_k_inv @ (x - mu_k)
        term3 = np.log(pi[direction])

        scores[direction] = term1 + term2 + term3

    return max(scores, key=scores.get)

y_pred_qda = x_test.apply(lambda row: qda_score(row, mu_train, sigma_qda, pi), axis=1)

conf_matrix = confusion_matrix(y_test, y_pred_qda)
accuracy = accuracy_score(y_test, y_pred_qda)

print("\nConfusion Matrix:")
print(pd.DataFrame(conf_matrix, 
                   index=['Actual Down', 'Actual Up'], 
                   columns=['Predicted Down', 'Predicted Up']))
print(f"\nAccuracy: {accuracy:.4f}")
```

Make plots of the training classification error and the test classification error as a function of the number of neighbors $K$ (or $1/K$; if you use $1/K$, make sure the x-axis is on the log scale). Which $K$ gives the best performance on the training data? On the test data? Which $K$ would you use?**

```python
from sklearn.neighbors import KNeighborsClassifier

K_values = range(1, 51)
train_errors = []
test_errors = []

for k in K_values:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)
    # Training error
    train_pred = knn.predict(X_train)
    train_error = 1 - accuracy_score(y_train, train_pred)
    train_errors.append(train_error)
    # Test error
    test_pred = knn.predict(X_test)
    test_error = 1 - accuracy_score(y_test, test_pred)
    test_errors.append(test_error)

# Plot errors vs 1/K (log scale)
plt.figure(figsize=(10, 6))
plt.plot(1 / np.array(K_values), train_errors, 'b-', label='Training Error')
plt.plot(1 / np.array(K_values), test_errors, 'r-', label='Test Error')
plt.xscale('log')
plt.xlabel('1/K (log scale)')
plt.ylabel('Misclassification Rate')
plt.legend()
plt.title('KNN Classification Error vs 1/K')
plt.show()

# Find best K for test data
best_test_k = K_values[np.argmin(test_errors)]
best_test_error = np.min(test_errors)
print(f"Best K on test data: {best_test_k}")

# Find best K for training data
best_train_k = K_values[np.argmin(train_errors)]
best_train_error = K_values[np.argmin(train_errors)]
print(f"Best K on training data: {best_train_k}")
```
